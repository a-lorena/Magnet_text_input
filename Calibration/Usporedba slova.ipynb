```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
fileA = "LetterA.json"
fileB = "LetterB.json"

# 1. slovo --> lokacija 1, iznad
# 2. slovo --> lokacija 1, pored
# 3. slovo --> lokacija 2, iznad
# 4. slovo --> lokacija 2, pored
```

```python
x_array_A = []
y_array_A = []
z_array_A = []

with open(fileA, 'r') as file:
    data = json.load(file)
    
    i = 1
    for letter in data:
        print("Number of points in letter " + str(i) + ": " + str(len(letter["fingerprint"])) + "\n")
        i += 1
        
        x_values = []
        y_values = []
        z_values = []
        l = letter["letter"]
        
        for point in letter["fingerprint"]:
            x_values.append(point[0])
            y_values.append(point[1])
            z_values.append(point[2])
            
        x_array_A.append(x_values)
        y_array_A.append(y_values)
        z_array_A.append(z_values)
```

```python
x_array_B = []
y_array_B = []
z_array_B = []
        
with open(fileB, 'r') as file:
    data = json.load(file)
    
    i = 1
    for letter in data:
        print("Number of points in letter " + str(i) + ": " + str(len(letter["fingerprint"])) + "\n")
        i += 1
        
        x_values = []
        y_values = []
        z_values = []
        l = letter["letter"]
        
        for point in letter["fingerprint"]:
            x_values.append(point[0])
            y_values.append(point[1])
            z_values.append(point[2])
            
        x_array_B.append(x_values)
        y_array_B.append(y_values)
        z_array_B.append(z_values)
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [16, 5]

print("USPOREDBA - LOKACIJA 1, MAGNET IZNAD UREƒêAJA")

values_A = []
for i in range(0, len(x_array_A[0])):
    values_A.append(i)
    
values_B = []
for i in range(0, len(x_array_B[0])):
    values_B.append(i)

    
figure, axis = plt.subplots(1, 3)

axis[0].plot(values_A, x_array_A[0])
axis[0].plot(values_B, x_array_B[0])
axis[0].set_title("X Values")
axis[0].set_xlabel("n")
axis[0].set_ylabel("value")
axis[0].legend(['A', 'B'])

axis[1].plot(values_A, y_array_A[0])
axis[1].plot(values_B, y_array_B[0])
axis[1].set_title("Y Values")
axis[1].set_xlabel("n")
axis[1].set_ylabel("value")
axis[1].legend(['A', 'B'])

axis[2].plot(values_A, z_array_A[0])
axis[2].plot(values_B, z_array_B[0])
axis[2].set_title("Z Values")
axis[2].set_xlabel("n")
axis[2].set_ylabel("value")
axis[2].legend(['A', 'B'])

yabs_max = abs(max(axis[0].get_ylim(), key=abs))
axis[0].set_ylim(ymin=-yabs_max, ymax=yabs_max)
yabs_max = abs(max(axis[1].get_ylim(), key=abs))
axis[1].set_ylim(ymin=-yabs_max, ymax=yabs_max)
yabs_max = abs(max(axis[2].get_ylim(), key=abs))
axis[2].set_ylim(ymin=-yabs_max, ymax=yabs_max)

#plt.savefig("Slova A i B iznad ureƒëaja.png", bbox_inches='tight')
plt.show()
```
