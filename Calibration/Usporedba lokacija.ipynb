```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
fileName = input()

#fileName --> LetterA.json ili LetterB.json

# 1. slovo --> lokacija 1, iznad
# 2. slovo --> lokacija 1, pored
# 3. slovo --> lokacija 2, iznad
# 4. slovo --> lokacija 2, pored
```

```python
x_array = []
y_array = []
z_array = []

x_array_z = []
y_array_z = []
z_array_z = []

x_array_s = []
y_array_s = []
z_array_s = []

with open(fileName, 'r') as file:
    data = json.load(file)
    
    i = 1
    for letter in data:
        print("Number of points in letter " + str(i) + ": " + str(len(letter["fingerprint"])) + "\n")
        i += 1
        
        x_values = []
        y_values = []
        z_values = []
        l = letter["letter"]
        
        for point in letter["fingerprint"]:
            x_values.append(point[0])
            y_values.append(point[1])
            z_values.append(point[2])
            
        x_array.append(x_values)
        y_array.append(y_values)
        z_array.append(z_values)
        
x_array_z.append(x_array[0])
x_array_z.append(x_array[2])
y_array_z.append(y_array[0])
y_array_z.append(y_array[2])
z_array_z.append(z_array[0])
z_array_z.append(z_array[2])

x_array_s.append(x_array[1])
x_array_s.append(x_array[3])
y_array_s.append(y_array[1])
y_array_s.append(y_array[3])
z_array_s.append(z_array[1])
z_array_s.append(z_array[3])
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [16, 5]

print("USPOREDBA - SLOVO " + l + ", MAGNET IZNAD UREƒêAJA")

values_1 = []
for i in range(0, len(x_array_z[0])):
    values_1.append(i)
    
values_2 = []
for i in range(0, len(x_array_z[1])):
    values_2.append(i)

    
figure, axis = plt.subplots(1, 3)

axis[0].plot(values_1, x_array_z[0])
axis[0].plot(values_2, x_array_z[1])
axis[0].set_title("X Values")
axis[0].set_xlabel("n")
axis[0].set_ylabel("value")
axis[0].legend(['Lokacija 1', 'Lokacija 2'])

axis[1].plot(values_1, y_array_z[0])
axis[1].plot(values_2, y_array_z[1])
axis[1].set_title("Y Values")
axis[1].set_xlabel("n")
axis[1].set_ylabel("value")
axis[1].legend(['Lokacija 1', 'Lokacija 2'])

axis[2].plot(values_1, z_array_z[0])
axis[2].plot(values_2, z_array_z[1])
axis[2].set_title("Z Values")
axis[2].set_xlabel("n")
axis[2].set_ylabel("value")
axis[2].legend(['Lokacija 1', 'Lokacija 2'])

yabs_max = abs(max(axis[0].get_ylim(), key=abs))
axis[0].set_ylim(ymin=-yabs_max, ymax=yabs_max)
yabs_max = abs(max(axis[1].get_ylim(), key=abs))
axis[1].set_ylim(ymin=-yabs_max, ymax=yabs_max)
yabs_max = abs(max(axis[2].get_ylim(), key=abs))
axis[2].set_ylim(ymin=-yabs_max, ymax=yabs_max)

#plt.savefig("Slovo " + l + " iznad ureƒëaja.png", bbox_inches='tight')
plt.show()
```
